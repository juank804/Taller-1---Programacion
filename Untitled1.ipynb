```python
# -*- coding: utf-8 -*-
# Leer un número entero positivo y determinar la suma de sus dígitos pares. 
#%%defino funciones
def ingresar_entero(): #para el ingreso de los valores
    while True:
        try:
            num = int(input("ingresar numero entero: "))
            if num>0: #me lo limita para que sea positvo
                break
        except:
            print("Numero no valido")
    return num

def separar_numeros(num): #vuelvo a utilizar la misma funcion
    lista=[]
    ult_dig = 0
    while num > 0:
        ult_dig = num % 10
        lista.append(ult_dig)
        num = num //10
    return lista

def suma_pares(lista):#funcion que me sumara los pares 
    pares=0  #contador que contendra los numeros pares y sumarlos
    for i in lista: #miro los elementos de la lista con los digitos
        if i %2==0: #si el residuo de un elemento al dividirlo entre 2 es 0
                    #me dara un numero par
            pares += i #este elemento si cumple se ira sumando 
    return pares #me dara la sumatoria de los numeros pares
#%%lectura de funciones
numero=ingresar_entero() #me arroja el numero ingresado
separado=separar_numeros(numero) #me lo separa en una lista
suma=suma_pares(separado) #me suma los pares teniendo como parametro la lista
#%%impresion de las funcioes
print('la suma de los numeros pares es:',suma) #me imprime el resultado
```

```python
# -*- coding: utf-8 -*-
#%%Leer un número entero positivo y determine si existen en el número dígitos 
#repetidos. Reporte cuales son los dígitos repetidos.
#%%Funciones
def entrada_numeros(): #entrada de numeros enteros
    while True:
        try:
            num=int(input('ingrese el numero: '))
            if num>0: #me lo limita a que sea positivo como es indicado
                break
        except:
            print('valor invalido entero')
    return num

def separar_numeros(num): #utilizo la misma funcion para pasarlo a una lista
    lista=[]
    ult_dig = 0
    while num > 0:
        ult_dig = num % 10
        lista.append(ult_dig)
        num = num //10
    return lista

def num_repetidos(lista):#la funcion que me mira los numeros repetidos
    repetidos=[] #lista vacia que contendra los numeros repetidos
    unicos=[] #lista vacia que me contendra los numeros unicos
    for i in lista: #me mira cada elemento de la lista de numeros
        if i not in unicos: #si el elemento no es unico
            unicos.append(i) #agrego ese elemento a unicos
        else: #si es falso
            if i not in repetidos: #es porque esta repetido y la idea es
                repetidos.append(i) #agregarlo a la lista de repetidos
    return repetidos #solo me interesa saber cuales son repetidos
#%%lectura de funciones
numeros=entrada_numeros() #numeros ingresados
separar=separar_numeros(numeros) #me lo convierte en una lista
repetidos=num_repetidos(separar) #me devuelve los numeros repetidos en la lista
#%%imprimir
print('repetidos son :',repetidos)
```

```python
# -*- coding: utf-8 -*-
#%%Iterar a través de los primeros cien enteros positivos, buscando los 
#múltiplos de 3 e imprimiendo y almacenándolos en una lista hasta encontrar 
#los primeros 15 de ellos. Una vez encontrados, continuar iterando en busca 
#de los múltiplos de 4 y almacenarlos en otra lista.
#%%funciones
def cien_num(): #funcion que dara los primeros 100 numeros desde el 0
    cien=[] #lista vacia que se llenara
    for i in range(0,101): #los elementos del 0 al 101 sin contar el 101
        cien.append(i) #los agregara a la lista llamada cien
    return cien #me dara los numeros

def multiplos_tres(cien): #funcion que me dara los multiplos de 3
    multiplos_3=[] #lista vacia que se llenara con los multiplos de 3
    for i in cien: #mirara todos los elementos de la lista
        if 0<i<47: #lo restrinjo hasta donde se encontrara los 15 primeros
                   #multiplos de 3
            if i%3==0: #el residuo entre la division de un numero entre 3 da 
                       #0 entonces es multiplo de 3
                multiplos_3.append(i) #lo agrego a la lista
    return multiplos_3

def multiplos_cuatro(cien): #funcion que me dara los multiplos de 4
    multiplos_4=[] #lista vacia que tendra los multiplos de 4
    for i in cien: #mira todos los elementos de la lista
        if 47<i<101: #lo restrinjo desde donde se acaba los primeros 15 
            #multiplos de 3 y pienso determinar el resto pero con 4
            if i%4==0: # me determina que sea multiplo de 4
                multiplos_4.append(i) #lo agrego a la lista
    return multiplos_4
#%%lectura de funciones
numeros=cien_num()
tres=multiplos_tres(numeros) 
cuatro=multiplos_cuatro(numeros)
#%%impresion
print('multiplos del 3 son:',tres)
print('multiplos del 4 son:',cuatro)
#%%
```

```python
# -*- coding: utf-8 -*-
#%%Leer una lista de números enteros y enumerar cuantos elementos se repiten
#exactamente dos veces. Utilizar para tal fin el método count() de listas.
#%%funciones
def ingresar_entero(): #para el ingreso de los valores
    while True:
        try:
            num = int(input("ingresar numero entero: "))
            if num>0: #me lo limita para que sea positvo
                break
        except:
            print("Numero no valido")
    return num

def separar_numeros(num): #vuelvo a utilizar la misma funcion
    lista=[]
    ult_dig = 0
    while num > 0:
        ult_dig = num % 10
        lista.append(ult_dig)
        num = num //10
    return lista

def si_dos_veces(lista):#funcion que tiene como pararmetro la lista de digitos
    cont=0 #contador que inicia en 0
    for i in lista: #miraremos los elementos de la lista
        repetido=lista.count(i)#cuanta cuantas veces esta repetidos un elemento
                               #en la lista
        if repetido==2:  #la restrinjo a que esten repetidos dos veces
            cont+= 1 #si pasa eso que me sume 1 al contador 
    return cont  #me retorne el contador
#%%lectura de funciones
numero=ingresar_entero() #variable que contiene el numero ingrasado
separar=separar_numeros(numero) #variable que contiene la lista con los digitos
repetido=si_dos_veces(separar) #me cuenta cuantos numeros hay repetidos
#%%imprimir
print('la cantidad de numeros que estan exactamente dos veces es: ',(repetido/2))
#en este caso se dividio entre dos,ya que me cuenta por cada numero repetido
#los dos, entonces lo dividimos para que me cuente la pareja ordenada como 1
#%%
```

```python
# -*- coding: utf-8 -*-
#%%Leer un número entero y determinar si la suma de sus dígitos es un número 
#de Fibonacci
#%% defino las funciones
def entrada_numeros(): #ingreso de numero entero
    while True:
        try:
            num=int(input('ingrese el numero: '))
            break
        except:
            print('valor invalido')
    return num
    
def sum_elementos(num): #me sumara cada elemento del numero que agrege
    suma=0 #el contador empieza en cero
    while num > 0: #mientras el numero sea positivo no se rompe el ciclo
        ult = num % 10 #me dara el resultado del ultimo digito
        suma = suma + ult 
        num = num // 10
    return suma

def sucesion_fibonacci(suma): #definir la sucesion de fibonacci 
                              #en funcion de la suma de los digitos
    fibonacci=[] #lista vacia que se ira llenando con los numeros de fibonacci
    a=0 #primer valor de fibonacci
    fibonacci.append(a) #se agregara a la lista
    b=1 #segundo valor de fibonacci
    fibonacci.append(b) #se agregara a la lista
    cont=1 #contador que empiece en 1
    for i in fibonacci: #mirare los elementos de la lista
        if cont<=suma:#mientras el contador sea menos a la suma de los digitos
            siguiente= fibonacci[cont] #la posicion del siguiente 
                                       #valor de fibonacci 
            el_que_sigue= i+siguiente #el siguiente valor de fibonacci
            fibonacci.append(el_que_sigue) #se ira agregando el valor que sigue 
            cont +=1 #y el contador ira aumentando
        else:
            break 
    return fibonacci
```
